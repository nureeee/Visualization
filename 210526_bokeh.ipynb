```python
import numpy as np
import pandas as pd
import seaborn as sns
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
p = figure(plot_width=600, plot_height=400)
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
p.circle(x, y, size=20)
show(p)
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
p.circle(x, y, size=20, line_color = 'pink', fill_color='m', fill_alpha=0.7)
show(p)
```
