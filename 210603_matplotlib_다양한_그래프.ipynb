# 1.Scatter

1-1. x, y, colors, area 설정하기
* colors : 어떠한 정수 값을 색상으로 표현해 줍니다.
* area : 점의 크기, 값이 커지면 점이 커진다.


```python
x= np.random.rand(50)
y= np.random.rand(50)

colors = np.arange(50)
area = x * y * 1000
```

```python
plt.scatter(x, y, s=area, c=colors) # s: 점의 크기, C :점의 색상
plt.show()
```

## 1-2 cmap, alpha
* cmap에 컬러를 지정하면, 컬러 값을 모두 동일하게 설정
* alpha는 투명도를 나타내며, 0~1 사이의 값을 갖고 0에 가까울수록 투명해진다

```python
# 현재 사용할 figure에 대해서만 크기 조절
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.scatter(x, y, s=area, c='b', alpha=0.2)
plt.title('alpha=0.2')

plt.subplot(132)
plt.scatter(x, y, s=area, c='r', alpha=0.5)
plt.title('alpha=0.5')

plt.subplot(133)
plt.scatter(x, y, s=area, c='g', alpha=1.0)
plt.title('alpha=1.0')

plt.colorbar() # 값졀 농도 확인하기

plt.show()
```

#3. line plot

## 3-1 기본 lineplot 그리기 

```python
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)
plt.grid()
plt.show()
```

## 3-2 2개 이사의 그래프 그리기
-컬러, 투명도 추가

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, color='blue', alpha=0.3, label='1 + sin')
plt.plot(x, y_2, color='red', alpha=0.7, label='1 + cos')

plt.grid()
plt.legend()
plt.show()
```

## marker 추가

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, color='blue', alpha=0.3, label='1 + sin', marker='o')
plt.plot(x, y_2, color='red', alpha=0.7, label='1 + cos', marker='+')

plt.grid()
plt.legend()
plt.show()
```

# 5.Histogram

5-1 기본 Histogram

```python
N = 100000
bins = 30 #십만개 데이터를 30개씩 끊어서 본다

x = np.random.randn(N) 

plt.hist(x, bins=bins) # y축이 필요 없는 이유는 히스토그램은 기본적으로 y축이 구간별 count를 의미하기 때문
plt.show()
 
```

## 5-2 다중 히스토그램 그려보기

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 3,
                         sharey=True,  # y축 그래프를 다중 그래프가 공유
                         tight_layout=True) # graph는 패딩을 자동으로 조절해서

fig.set_size_inches(12, 5)

axes[0].hist(x, bins=bins)
axes[1].hist(x, bins=bins * 2)
axes[2].hist(x, bins=bins * 4)

plt.show()
```

## 5-3 y축에 Density 표기

```python
N = 100000
bins = 30
x = np.random.randn(N)

fig, axes = plt.subplots(1, 2,
                         tight_layout=True)

fig.set_size_inches(9, 3)

# density=True 값을 통하여 Y축에 Density를 표기할 수 있다.
axes[0].hist(x, bins=bins, density=True, cumulative=True) # cumulative : 빈도수 누적합
axes[1].hist(x, bins=bins, density=True)

plt.show()
```

## pie chart
* explode : 파이에서 툭 튀어져 나온 비율
* autopct : 펀센트 자동으로 표기
* shadow : 

```python
labels =['a', 'b', 'c', 'd']
sizes =[12.2, 22, 32.3, 41]
explode = (0.3, 0, 0, 0)

_, texts, autotexts = plt.pie(sizes,
                              labels=labels,
                              explode=explode,
                              autopct='%1.1f%%',
                              shadow=True,
                              startangle=90
                              )
plt.title('pie chart test')

# label텍스트에 개한 스타일 효과
for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')

# pie 위읭 덱스트에 대한 스타일 적용
for i in autotexts:
    t.set_color('white')
    t.set_fontsize(20)

plt.show()
```

## Box Plot

```python
# 기본 데이터 분포
spread = np.random.rand(50) * 100

#강제로 중앙값 
center = np.ones(25) * 50

# 윗 쪽 아웃라이어 만들기 
filier_high = np.random.rand(10) * 100 + 100 # 최소100 , 최대200

# 아래 쪽 아웃라이어 만들기
filier_low = np.random.rand(10) * -100 #최소-100 최대0
# 모든 데이터 합치기
data = np.concatenate([spread, center, filier_high, filier_low])
len(data)

plt.hist(data)
```

```python
plt.boxplot(data)
plt.show()
```

## 7-2 다중 박스 프롯 생성

```python
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
filier_high = np.random.rand(10) * 100 + 100 
filier_low = np.random.rand(10) * -100 
data = np.concatenate([spread, center, filier_high, filier_low])

spread = np.random.randn(50) * 100
center = np.ones(25) * 40 # 중앙값을 약간 내려보기

filier_high = np.random.rand(10) * 100 + 100 
filier_low = np.random.rand(10) * -100 

d2 = np.concatenate([spread, center, filier_high,filier_low])
```

```python
data.shape, d2.shape
```

```python
data_re = data.reshape(-1, 1)
d2_re = d2.reshape(-1, 1)
final_data = [data_re, d2_re, d2_re[::2, 0]]
```

```python
plt.boxplot(final_data)
plt.show()
```

```python
plt.boxplot(final_data, vert=False)
plt.show()
```

```python
outlier_marker = {'markerfacecolor':'r', 'marker':'d'}
plt.boxplot(final_data, flierprops=outlier_marker)
plt.show()
```
