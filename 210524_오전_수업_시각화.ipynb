```python
x = np.arange(1, 11)
y1 = x**2
y2 = x**3
y3 = x**4
x, y1, y2, y3
```

```python
color = ['pink', 'olive', 'orange', 'skyblue']
hobby = ['여행', '독서', '자전거', '게임']

numbers = [20, 5, 7, 18]

plt.figure(figsize=(6, 6))

plt.pie(numbers, labels=hobby, colors=color, autopct='%.1f%%')  ## 소수점이하 자리 표시

plt.axis('equal')
plt.title('취미 비율')

plt.show()
```

```python
np.random.seed(100)
d = np.random.randint(1, 100,(3, 6))
x= np.arange(6)

for i, c in zip (range(len(d)), list('gbr')):
    plt.bar(x + i*0.2, d[i], width=0.2)

plt.title('상반기 과일 판매량')

plt.legend(labels=['사과', '배', '포도'])
plt.xticks(x+0.2, labels=['1월', '2월', '3월', '4월', '5월', '6월', ])
plt.yticks(np.arange(0, 100, 10))

plt.show()
```
