# 1. 밑그림 그리기

```python
# 데이터 생성
data = np.arange(1, 100)

# 그리기
plt.plot(data)

#그래프 보여주기(노트북에서는 생략이 가능하지만 써주기)
plt.show()

# 아래 그래프는 캠버스 하나에 하나의 내용만 있는 단일 그래프
```

# 1-2 다중 그래프 (multiple graphs)
 1개의 캔버스 안에 여러 개의 그래프를 그리는 것


```python
data1 = np.arange(1, 51)
data2 = np.arange(51, 101)

# 그래프를 
plt.plot(data1)
plt.plot(data2)

# show 전에 여러개의 그래프를 동시에 올릴 수 있다.
plt.show()
```

2개의 figure로 나누어서 다중 그래프 그리기
* `plt.figure()` : 캔버스 설정함수, 호출 할 때마다 캔버스가 하나씩 만들어 진다.

```python
plt.plot(data1) # data1 시각화

plt.figure() # 새로운 캔버스 생성

plt.plot(data2) # 새롭게 만들어진 캔버스에 data2 시각화

plt.show() # 모두 보여주기
```

# 1-3 여러개의 plot을 그리는 방법(subplot)
* `subplot(row, column, index)` 여기서 `index`는 1부터 시작

```python
plt.subplot(2, 1, 1) # 2개의 행을 만들고 하나의 열을 만들어서 첫번째 plot에 표시
plt.plot(data1)

plt.subplot(2, 1, 2)
plt.plot(data2)

plt.show()
```

위와 똑같은 그래프를 그리지만 다른 표현
* 콤마(,)를 생략하고 그릴 수도 있다

```python
plt.subplot(211)  ## row:2 columns:1 index:1
plt.plot(data1)

plt.subplot(212)
plt.plot(data2)

plt.show()
```

```python
plt.subplot(121)  
plt.plot(data1)

plt.subplot(122)
plt.plot(data2)

plt.show()
```

## 1-4 여러 plot을 그리는 방법(subplot) -s가 저 붙었음
* 'figure', axes = plt.subplots(행의 개수, 열의 개수)
* figure : 캔버스 설정을 할 수 있는 객체
*


```python
fig, axes = plt.subplots(2, 3)

axes[0, 0].plot(data1)
axes[0, 1].plot(data1 * data1)
axes[0, 2].plot(data1 ** 3)

axes[1, 0].plot(data2 + data1)
axes[1, 1].plot(-3 * data2)
axes[1, 2].plot(data2 + 2)

plt.tight_layout()
plt.show()
```

## 스타일 세부 설정 marker, line, color 설정
**marker의 종류**
* '.'	point marker
* ','	pixel marker
* 'o'	circle marker
* 'v'	triangle_down marker
* '^'	triangle_up marker
* '<'	triangle_left marker
* '>'	triangle_right marker
* '1'	tri_down marker
* '2'	tri_up marker
* '3'	tri_left marker
* '4'	tri_right marker
* 's '	square marker
* 'p'	pentagon marker
* '*'	star marker
* 'h'	hexagon1 marker
* 'H'	hexagon2 marker
* '+'	plus marker
* 'x'	x marker
* 'D'	diamond marker
* 'd'	thin_diamond marker
* '|'	vline marker
* '_'	hline marker

```python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', markersize=5)
plt.plot(data, data * 3, marker='D', markersize=10)
plt.plot(data, data * 2 -3, marker='*', markersize=15)
plt.plot(data, data * 2 -5, marker='x', markersize=10)

plt.show()
```

## line 종류
* '-' solid line style
* '--' dashed line style
* '-.' dash-dot line style
* ':' dotted line style


```python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', linestyle='') # 선이 안보임
plt.plot(data, data * 3, marker='D', linestyle='-')
plt.plot(data, data * 2 -3, marker='*', linestyle='--')
plt.plot(data, data * 2 -2, marker='x', linestyle='-.')
plt.plot(data, data * 2 -5, marker='x', linestyle=':')

plt.show()
```

## color

```python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', linestyle='', color='y') # 선이 안보임
plt.plot(data, data * 3, marker='D', linestyle='-', color='m')
plt.plot(data, data * 2 -3, marker='*', linestyle='--', color='c')
plt.plot(data, data * 2 -2, marker='x', linestyle='-.', color='r')
plt.plot(data, data * 2 -5, marker='x', linestyle=':', color='b')

plt.show()
```

alpha를 사용하면 투명도 조절이 가능
* 0 ~ 1사이로 설정
* 0에 가까울수록 투명해지고, 1에 가까울수록 불투명해 진다.

```python
data = np.arange(10)

plt.plot(data, data * 2, marker='o', linestyle='', alpha=0.1) # 선이 안보임
plt.plot(data, data * 3, marker='D', linestyle='-', alpha=0.3)
plt.plot(data, data * 2 -3, marker='*', linestyle='--', alpha=0.9)
plt.plot(data, data * 2 -2, marker='x', linestyle='-.', alpha=0.9)
plt.plot(data, data * 2 -5, marker='x', linestyle=':', alpha=0.9)

plt.show()
```
